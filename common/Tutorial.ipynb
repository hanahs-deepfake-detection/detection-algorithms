# How to use `dataset_util`

```python
import cv2 as cv
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patch
import dlib
```

Load the dataset.

```python
import sys
sys.path.insert(0, '../common/')
from dataset_util import Dataset

dataset = Dataset()
dataset.get_metadata_dataframe()
```

Let's take a look at some videos.

```python
video_filename = 'aagfhgtpmv.mp4'
plt.imshow(cv.cvtColor(dataset.get_frame_from_video(video_filename, 0), cv.COLOR_BGR2RGB))
plt.show()
```

Extract faces.

```python
for frame_num in range(10):
    ax = plt.subplot(2, 5, frame_num + 1)
    face = dataset.get_face_from_video(video_filename, frame_num)
    ax.imshow(cv.cvtColor(face, cv.COLOR_BGR2RGB))
plt.show()
```
