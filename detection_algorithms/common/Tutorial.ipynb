# Tutorial of `dataset_util`
This notebook covers how to use `dataset_util`. Import first.

```python
from dataset_util import Dataset
```

Use a pre-downloaded faceforensics dataset.

```python
dataset = Dataset('../input/faceforensics/', 'faceforensics')
```

Take a look at the metadata dataframe.

```python
df = dataset.get_metadata_dataframe()
df
```

Fetch the 0th frame from file `469.mp4`.

```python
file = '../input/faceforensics/original_sequences/actors/raw/videos/13__walking_down_indoor_hall_disgust.mp4'
frame = dataset.get_frame_from_video(file, 0)
frame.shape
```

Plot the frame.

```python
plt.imshow(cv.cvtColor(frame, cv.COLOR_BGR2RGB))
plt.show()
```

Get face coords.

```python
coords = dataset.get_face_coords(file, 0)
coords
```

Plot the face only.

```python
plt.imshow(cv.cvtColor(dataset.crop_frame(file, 0, coords), cv.COLOR_BGR2RGB))
plt.show()
```
